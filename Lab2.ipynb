### Missing Values

2. Choose Strategy

Case A: Small gaps (few missing rows in a continuous timeline)

Use forward fill (ffill) and backward fill (bfill):

```
df.fillna(method="ffill", inplace=True)  # fill forward
df.fillna(method="bfill", inplace=True)  # fill backward if start is NaN

```

Case B: Larger missing chunks (but still reasonable)

Impute with rolling median/mean to preserve local trends:


```
df['Global_active_power'] = df['Global_active_power'].fillna(
    df['Global_active_power'].rolling(window=10, min_periods=1, center=True).median()
)

```

Case C: Entire column has large missing proportion

If >30–40% missing, either:

Drop that column (if not crucial), or

Use advanced imputation (KNN imputer, interpolation).



```
df.interpolate(method="time", inplace=True)
```







##3. Exploratory Data Analysis (EDA)

````python
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')
df['Global_active_power'] = df['Global_active_power'].replace([np.inf, -np.inf], np.nan)

# Plotting histogram
plt.figure(figsize=(10,6))
plt.hist(df['Global_active_power'].dropna(), bins=50, color='g', edgecolor='black')
plt.title('Histogram of Global Active Power')
plt.xlabel('Global Active Power (kW)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

This histogram provides a visual representation of the distribution of "Global Active Power," which reflects the electricity consumption in kilowatts (kW). From the graph, we can observe that the majority of the data points correspond to lower power consumption values, close to 0 kW, indicating that most instances of the data set involve minimal power usage. As we move towards higher values, the frequency of data points decreases significantly, suggesting that large power consumption events are less common. The histogram is plotted with 50 bins, helping to clearly delineate the frequency of each range of power consumption. Additionally, any infinite or missing values have been removed to ensure the data is clean and accurately represents the underlying trends.


````python
# Selecting numeric columns
numeric_data =df[['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity',
                     'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']]

# Plotting correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
````

This heatmap visualizes the correlation between various numeric variables, such as "Global Active Power," "Voltage," and "Sub Metering" values. Strong positive correlations, such as between "Global Active Power" and "Sub Metering 3" (0.64), are shown in red, while negative correlations, like between "Voltage" and "Global Active Power" (-0.40), are in blue. The heatmap provides a clear overview of how different variables relate to one another, which can help identify patterns in energy consumption and meter readings.

````python
# Plotting Power Consumption by Month and Day of the Week
plt.figure(figsize=(10, 6))

# Get the groupby count and reset the index
grouped_data = df.groupby(['Month', 'Day_of_Week']).size().reset_index(name='Frequency')

# Create a new column for month-day combination
grouped_data['Month_Day'] = grouped_data['Month'].astype(str) + '-' + grouped_data['Day_of_Week'].astype(str)

# Plotting
plt.plot(grouped_data['Month_Day'], grouped_data['Frequency'], label='Power Consumption Frequency', color='b')

plt.title('Power Consumption by Month and Day of the Week')
plt.xlabel('Month and Day of Week')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability
plt.legend()
plt.grid(True)
plt.show()
````



The plot displays the frequency of power consumption occurrences across different months and days of the week. It reveals fluctuations in power consumption, with noticeable peaks and valleys across various combinations of time.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the violin plot for 'Global_active_power' by month
plt.figure(figsize=(10, 6))
sns.violinplot(x='Month', y='Global_active_power', data=df, palette='coolwarm')
plt.title('Global Active Power Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Global Active Power (kW)')
plt.grid(True)
plt.show()
````

This Violin Plot provides a clear view of how Global Active Power is distributed across each month. Each "violin" shape shows the range of power consumption, with wider sections indicating where most of the values lie. The white dot in the center represents the median, while the thick black lines show the range between the 25th and 75th percentiles. From the plot, we can see that power consumption is fairly consistent throughout the year, with some months showing slightly higher concentrations of power usage.
